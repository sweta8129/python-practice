# 12 File Formats
## Why it matters
Choosing the right format (CSV/JSON/Parquet/Avro) affects speed, cost, and schema control.

## Concepts
- Row vs column storage
- Compression and schema
- When to use Parquet in analytics pipelines


```python
import pandas as pd

df = pd.DataFrame({"id": [1, 2], "value": [100, 200]})
df.to_csv("sample.csv", index=False)
df.to_parquet("sample.parquet", index=False)
print("Wrote CSV and Parquet examples")
```
