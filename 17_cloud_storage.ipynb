# 17 Cloud Storage
## Why it matters
Data lakes depend on object storage such as S3 and Azure Blob.

## Concepts
- Buckets/containers, prefixes, object keys
- Upload/download/list operations
- Reading Parquet from cloud paths


```python
s3_example = '''
import boto3
s3 = boto3.client("s3")
s3.upload_file("sample.parquet", "my-bucket", "raw/sample.parquet")
'''

azure_example = '''
from azure.storage.blob import BlobServiceClient
service = BlobServiceClient.from_connection_string("<connection_string>")
'''

print(s3_example)
print(azure_example)
```
