## Patient class task from earlier notebook
This completes the prompt from your introductory notebook by creating reusable patient objects.

## Exercises
- Add validation for age
- Add method to convert to dictionary for JSON export


```python
from dataclasses import dataclass

@dataclass
class Patient:
    name: str
    age: int
    is_new: bool

    def summary(self):
        status = "new" if self.is_new else "existing"
        return f"{self.name} ({self.age}) - {status} patient"

p1 = Patient("John Smith", 20, True)
p2 = Patient("Mary Jane", 31, False)
print(p1.summary())
print(p2.summary())
```
