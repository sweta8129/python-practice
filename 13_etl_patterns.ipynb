# 13 ETL Patterns
## Why it matters
Reliable ETL requires idempotency, incremental logic, and data quality checks.

## Concepts
- ETL vs ELT
- Watermark-based incremental loads
- SCD Type 1 and Type 2
- Validation checks before load


```python
def incremental_filter(rows, last_ts):
    return [r for r in rows if r["updated_at"] > last_ts]

rows = [
    {"id": 1, "updated_at": "2026-04-19"},
    {"id": 2, "updated_at": "2026-04-20"}
]
print(incremental_filter(rows, "2026-04-19"))
```
