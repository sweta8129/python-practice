## Project steps

2. Transform data using Pandas
3. Save to Parquet and SQL table
4. Orchestrate with Airflow DAG
5. Add quality checks (nulls, duplicates, ranges)

## Interview outcome
Use this notebook as portfolio evidence and talking points in DE interviews.


```python
import pandas as pd

raw = pd.DataFrame([
    {"repo": "a", "stars": 10, "language": "Python"},
    {"repo": "b", "stars": 25, "language": "Python"}
])

transformed = raw.assign(star_band=lambda d: d["stars"].apply(lambda x: "high" if x >= 20 else "low"))
transformed.to_parquet("capstone_output.parquet", index=False)
print(transformed)
```
