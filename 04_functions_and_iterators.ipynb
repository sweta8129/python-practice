# 04 Functions and Iterators

## Why it matters
Reusable functions and generators are core for scalable ETL transformations.

## Core concepts
- `*args`, `**kwargs`
- Lambda and closures
- Generators and `yield`
- `itertools`

## Common pitfalls
- Mutable default arguments
- Exhausting a generator and reusing it

## Exercises
- Build a generator that yields records in chunks
- Create a decorator that times function execution


```python
def total(*nums):
    return sum(nums)

print(total(1, 2, 3, 4))

def chunked(items, size):
    for i in range(0, len(items), size):
        yield items[i:i + size]

for chunk in chunked(list(range(10)), 3):
    print(chunk)
```
