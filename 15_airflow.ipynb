# 15 Airflow
## Why it matters
Airflow orchestrates ETL jobs with scheduling, retries, and dependencies.

## Concepts
- DAG definition
- Operators (`PythonOperator`, `BashOperator`)
- Scheduling and retries


```python
example_dag = '''
from airflow import DAG
from airflow.operators.python import PythonOperator
from datetime import datetime

def run_task():
    print("ETL step")

with DAG("sample_de_dag", start_date=datetime(2026, 1, 1), schedule="@daily", catchup=False) as dag:
    task = PythonOperator(task_id="run_task", python_callable=run_task)
'''
print(example_dag)
```
