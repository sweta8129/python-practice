## Why it matters
Data engineers clean text fields, parse logs, and normalize messy IDs using string operations and regex.

## Core concepts

- f-strings
- `re.findall`, `re.sub`

## Common pitfalls
- Regex too broad/greedy
- Ignoring whitespace normalization

## Interview questions

## Exercises
- Extract all emails from sample text
- Normalize user-entered phone strings


```python
import re

text = "Contact: data.team@example.com and hr@example.org"
emails = re.findall(r"[\w\.-]+@[\w\.-]+", text)
print(emails)

raw = "  ETL_PIPELINE_2026  "
clean = raw.strip().lower().replace("_", "-")
print(clean)
```
