```python
# Solution 3
text = "data engineering data python python python"
freq = {}
for word in text.split():
    freq[word] = freq.get(word, 0) + 1
print(freq)
```

```python
# Solution 4
rows = [("u1", 10), ("u2", 5), ("u1", 7)]
res = {}
for user_id, amount in rows:
    res[user_id] = res.get(user_id, 0) + amount
print(res)
```

```python
# Solution 5
transactions = [
    {"id": 1, "customer_id": "C1"},
    {"id": 2, "customer_id": "C2"},
    {"id": 3, "customer_id": "C1"},
]
by_customer = {}
for tx in transactions:
    cid = tx["customer_id"]
    by_customer.setdefault(cid, []).append(tx["id"])
print(by_customer)
```
