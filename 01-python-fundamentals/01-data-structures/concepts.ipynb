# Week 1: Data Structures

## Why this matters for DE
Choosing the right structure affects speed, memory, and correctness in ETL pipelines.

## Lists, Tuples, Sets, Dicts
- List: ordered, mutable
- Tuple: ordered, immutable
- Set: unique values, fast membership checks
- Dict: key-value lookup in O(1) average time

```python
records = ["a", "b", "a", "c"]
print("list:", records)
print("unique using set:", set(records))
```

```python
event = ("user_1", "click", "2026-04-16")
print(event[0], event[1])
```

```python
counts = {}
for item in records:
    counts[item] = counts.get(item, 0) + 1
print(counts)
```

## Common pitfall
Avoid naming a variable `dict`, `list`, or `set`; it shadows built-in types.
