# Week 2: Advanced Functions

## Why this matters for DE
Reusable, composable functions reduce bugs in transformation pipelines.

```python
def summarize(*values, round_to=2):
    total = sum(values)
    avg = total / len(values)
    return {"total": round(total, round_to), "avg": round(avg, round_to)}

print(summarize(10, 20, 30, round_to=1))
```

```python
rows = [{"id": 1, "amount": 40}, {"id": 2, "amount": 10}]
rows_sorted = sorted(rows, key=lambda r: r["amount"], reverse=True)
print(rows_sorted)
```

```python
from functools import reduce
vals = [1, 2, 3, 4]
print(list(map(lambda x: x * 2, vals)))
print(list(filter(lambda x: x % 2 == 0, vals)))
print(reduce(lambda a, b: a + b, vals))
```

```python
def log_call(fn):
    def wrapper(*args, **kwargs):
        print(f"running {fn.__name__}")
        return fn(*args, **kwargs)
    return wrapper

@log_call
def transform(x):
    return x * 10

print(transform(5))
```
