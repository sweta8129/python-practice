# Solutions: Advanced Functions

```python
# Solution 1
def multiply_all(*nums):
    result = 1
    for n in nums:
        result *= n
    return result

print(multiply_all(2, 3, 4))
```

```python
# Solution 2
rows = [
    {"id": 1, "ts": "2026-04-16T10:20:00"},
    {"id": 2, "ts": "2026-04-15T12:00:00"},
]
print(sorted(rows, key=lambda r: r["ts"]))
```

```python
# Solution 3
records = [{"amount": 10}, {"amount": -1}, {"amount": 0}, {"amount": 8}]
valid = list(filter(lambda r: r["amount"] > 0, records))
print(valid)
```

```python
# Solution 4
import time

def timed(fn):
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        out = fn(*args, **kwargs)
        end = time.perf_counter()
        print(f"{fn.__name__} took {end - start:.6f}s")
        return out
    return wrapper

@timed
def heavy(n):
    return sum(range(n))

print(heavy(1_000_000))
```
