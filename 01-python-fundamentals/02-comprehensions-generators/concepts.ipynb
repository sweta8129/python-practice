# Week 1: Comprehensions and Generators

## Why this matters for DE
Generators let you stream large files without loading everything into memory.

```python
nums = list(range(10))
squares = [n * n for n in nums]
print(squares)
```

```python
squares_map = {n: n * n for n in range(5)}
print(squares_map)
```

```python
gen = (n * n for n in range(1_000_000))
print(next(gen))
print(next(gen))
```

```python
def read_lines_lazy(path):
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            yield line.strip()
```
