# Solutions: Comprehensions and Generators

```python
# Solution 1
squares = [n*n for n in range(1, 31)]
print(squares[:5], squares[-3:])
```

```python
# Solution 2
sentence = "stream large data efficiently"
lengths = {w: len(w) for w in sentence.split()}
print(lengths)
```

```python
# Solution 3
def evens_only(values):
    for v in values:
        if v % 2 == 0:
            yield v

print(list(evens_only(range(10))))
```

```python
# Solution 4
def fake_rows(total):
    for i in range(total):
        yield {"row_id": i, "value": i % 7}

count = 0
for _ in fake_rows(1_000_000):
    count += 1
print(count)
```
