# Solutions: File I/O and Context Managers

```python
# Solution 1
from pathlib import Path
path = Path("numbers.txt")
with path.open("w", encoding="utf-8") as f:
    for i in range(1, 101):
        f.write(f"{i}\n")
```

```python
# Solution 2
from pathlib import Path
path = Path("numbers.txt")
total = 0
with path.open("r", encoding="utf-8") as f:
    for line in f:
        total += int(line.strip())
print(total)
```

```python
# Solution 3
import json
records = [{"id": 1}, {"id": 2}]
with open("records.json", "w", encoding="utf-8") as f:
    json.dump(records, f)
with open("records.json", "r", encoding="utf-8") as f:
    print(json.load(f))
```

```python
# Solution 4
def grep_lines(path, keyword):
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            if keyword in line:
                yield line.strip()
```
