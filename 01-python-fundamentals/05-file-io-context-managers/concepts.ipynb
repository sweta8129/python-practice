# Week 2: File I/O and Context Managers

## Why this matters for DE
Pipelines read/write files constantly. `with` prevents leaked file handles and partial writes.

```python
from pathlib import Path

base = Path("tmp_data")
base.mkdir(exist_ok=True)
out = base / "sample.txt"

with out.open("w", encoding="utf-8") as f:
    f.write("line1\nline2\n")

with out.open("r", encoding="utf-8") as f:
    print(f.read())
```

```python
import json
from pathlib import Path

record = {"id": 1, "name": "Sweta"}
path = Path("tmp_data/record.json")
with path.open("w", encoding="utf-8") as f:
    json.dump(record, f)

with path.open("r", encoding="utf-8") as f:
    print(json.load(f))
```
