# Solutions: Error Handling

```python
# Solution 1
def to_int_or_default(value, default=-1):
    try:
        return int(value)
    except ValueError:
        return default

print(to_int_or_default("12"))
print(to_int_or_default("12.5"))
```

```python
# Solution 2
class SchemaError(Exception):
    pass

def validate_record(record, required_keys):
    missing = [k for k in required_keys if k not in record]
    if missing:
        raise SchemaError(f"missing keys: {missing}")
    return True

print(validate_record({"id": 1, "name": "A"}, ["id", "name"]))
```

```python
# Solution 3
import time

def retry_call(func, retries=3):
    for i in range(retries):
        try:
            return func()
        except Exception:
            if i == retries - 1:
                raise
            time.sleep(0.1)

state = {"calls": 0}
def unstable():
    state["calls"] += 1
    if state["calls"] < 2:
        raise RuntimeError("try again")
    return "success"

print(retry_call(unstable, retries=3))
```
