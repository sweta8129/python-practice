# Week 2: Error Handling

## Why this matters for DE
Data pipelines fail in production. Handle errors explicitly so jobs are observable and retryable.

```python
def parse_int(value):
    try:
        return int(value)
    except ValueError as exc:
        print(f"bad integer: {value}")
        raise exc

# parse_int("10.2")
```

```python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return None
    finally:
        print("divide attempted")

print(safe_divide(10, 2))
```

```python
import time

def retry(func, retries=3, wait=1):
    for attempt in range(1, retries + 1):
        try:
            return func()
        except Exception as e:
            print(f"attempt {attempt} failed: {e}")
            if attempt == retries:
                raise
            time.sleep(wait)

count = {"n": 0}
def flaky():
    count["n"] += 1
    if count["n"] < 3:
        raise RuntimeError("temporary failure")
    return "ok"

print(retry(flaky, retries=4, wait=0))
```
