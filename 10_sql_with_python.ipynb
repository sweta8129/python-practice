# 10 SQL with Python
## Why it matters
SQL is core for DE interviews and production analytics layers.

## Concepts
- `sqlite3` from Python
- JOINs and aggregations
- Window functions (`ROW_NUMBER`, `LAG`)
- Parameterized queries for safety


```python
import sqlite3
conn = sqlite3.connect(":memory:")
cur = conn.cursor()
cur.execute("CREATE TABLE sales(team TEXT, amount INT)")
cur.executemany("INSERT INTO sales VALUES(?, ?)", [("A", 100), ("A", 50), ("B", 75)])
for row in cur.execute("SELECT team, SUM(amount) total FROM sales GROUP BY team"):
    print(row)
conn.close()
```
