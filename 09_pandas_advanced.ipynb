# 09 Pandas Advanced
## Why it matters
Advanced joins, null handling, and datetime logic are used in almost every ETL job.

## Concepts
- `merge`, `join`, `concat`
- Missing data strategy
- Datetime transformations
- Performance tips: vectorization vs `apply`


```python
import pandas as pd
left = pd.DataFrame({"id": [1, 2, 3], "value": [10, None, 30]})
right = pd.DataFrame({"id": [2, 3], "category": ["x", "y"]})
out = left.merge(right, on="id", how="left")
out["value"] = out["value"].fillna(out["value"].median())
print(out)
```
