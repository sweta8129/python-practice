# 11 APIs and Ingestion
## Why it matters
Most pipelines begin by extracting data from HTTP APIs.

## Concepts
- `requests.get` and status codes
- Pagination loop pattern
- Retry and backoff basics
- Idempotent writes using unique keys


```python
import time
import requests

def fetch_with_retry(url, retries=3):
    for attempt in range(retries):
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            return response.json()
        time.sleep(2 ** attempt)
    raise RuntimeError("Failed after retries")

print("Function ready for ingestion tasks")
```
