## Exercises
- Compute z-score normalization for a numeric array
- Replace loop-based feature engineering with vectorization


```python
import numpy as np
arr = np.array([10, 20, 30, 40], dtype=float)
print("mean:", arr.mean())
print("normalized:", (arr - arr.mean()) / arr.std())
```
