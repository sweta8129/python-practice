# 14 PySpark
## Why it matters
Spark is common for large-scale transformations beyond a single machine.

## Concepts
- SparkSession
- Transformations vs actions
- Joins, windows, partitioning


```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local[*]").appName("de-learning").getOrCreate()
df = spark.createDataFrame([(1, "A", 100), (2, "A", 150), (3, "B", 200)], ["id", "team", "sales"])
df.groupBy("team").sum("sales").show()
spark.stop()
```
