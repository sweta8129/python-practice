## Why it matters

## Core concepts
- `try/except/else/finally`
- File handling with context manager
- JSON and CSV basics
- Modules and type hints

## Exercises

```python
import json

payload = '{"job": "daily_sales", "status": "ok"}'
try:
    obj = json.loads(payload)
except json.JSONDecodeError as exc:
    print("Invalid JSON:", exc)
else:
    print("Parsed:", obj)
finally:
    print("Done")
```
